## How many data points for each glacier? 

```python
def count_datapoints_per_glacier(df, id_ls):
    
    count_ls = []
    for rgi in id_ls:
        df_sub = df_full.loc[df_full['RGIId'] == rgi]
        count = df_sub['dis_mag_my'].count()
        
        count_ls.append(count)

    d = {'RGIId':id_ls,
         'count':count_ls}
    df = pd.DataFrame(d)
    return df
```

```python
count_df = count_datapoints_per_glacier(df_full, rgi_ids)
```

```python
count_df = count_df.merge(rgi_15_sub[['RGIId','Area']], how='inner', on = 'RGIId')
```

```python
fig, ax = plt.subplots()
count_df.plot(ax=ax, x='Area',y='count', marker='o',linestyle='None', title='# of "valid" centerline velocity measurements per glacier')
for idx, row in count_df.iterrows():
    if row['count'] < 100:
        ax.annotate(row['RGIId'], (row['Area'], row['count']), rotation=30)
```

```python
winter_df = df_full.loc[df_full['season'] == 'DJF'][['dis_mag_my','z','RGIId']]
```

```python
winter_df = winter_df.dropna()
```

```python
def pivot_and_recombine(season_df, rgi_ids):
    '''this is to prepare wide dataframes for heatmap
    '''
    
    wide_df_ls = []
    
    for glacier in rgi_ids:
        sub_df = season_df.loc[season_df['RGIId'] == glacier]
        sub_df_w = sub_df.pivot_table(index='RGIId', values='dis_mag_my',columns='z')
        wide_df_ls.append(sub_df_w)
    wide_df = pd.concat(wide_df_ls)
    
    return wide_df
```

```python
df_wide = pivot_and_recombine(winter_df, rgi_ids)
```
