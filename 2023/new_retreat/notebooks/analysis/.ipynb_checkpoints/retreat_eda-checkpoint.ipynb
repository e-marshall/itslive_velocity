```python
def combine_lists(list1, list2):
    combo_ls = [] 
    
    for element in range(len(list1)):
        
        for i in range(len(list2)):
            
        
            ls1 = [list1[element],list2[i]]
            combo_ls.append(ls1)
            
    return combo_ls
```

```python
full_df.loc[(full_df['season'] == 'winter') & (full_df['z'] == 'z0')& (full_df['TermType'] == 'land') & (full_df['var'] == 'dis_mag')]['magnitude'].mean()
```

```python
def make_barplot_vals(full_df, termtype):

    z_ls = ['z0','z1','z2','z3','full']
    season_ls = ['winter','spring','summer','fall']

    lvars, means = [],[]
    for z in z_ls:

        for season in season_ls:

            varname = f'{z}_{season}'
            lvars.append(varname)
            mean = full_df.loc[(full_df['season'] == season) & (full_df['var'] == 'dis_mag') & (full_df['TermType'] == termtype) & (full_df['z'] == z)]['magnitude'].mean()
            #print(varname, ': ', mean)
            means.append(mean)

        means_dict = dict(zip(lvars, means))
    
    return means_dict

```

```python
land_means = make_barplot_vals(full_df, 'land')
```

```python
lake_means = make_barplot_vals(full_df, 'lake')
```

```python
def add_seasonal_change_cols(input_df):
    
    input_df['winter_minus_fall'] = input_df['winter'] - input_df['fall']
    input_df['spring_minus_winter'] = input_df['spring'] - input_df['winter']
    input_df['summer_minus_spring'] = input_df['summer'] - input_df['spring']
    input_df['fall_minus_summer'] = input_df['fall'] - input_df['summer']
    
    return input_df
```

## Organizing into tidy format and plotting....

```python
def organize_single_df(df, term_str):
    
    df = df.drop('Unnamed: 0', axis=1)
    #add term type col
    df['TermType'] = [term_str for i in range(len(df['RGIId'].to_list()))]
    df = df.melt(id_vars=['RGIId','z','TermType'])
    df = df.rename({'variable':'season'},axis=1)
    df = df.rename({'value':'mean_velocity'},axis=1)
    
    return df
```

```python
land_df = organize_single_df(lands_df, 'Land')
lake_df = organize_single_df(lakes_df, 'Lake')
```

```python
df_full = pd.concat([land_df, lake_df])
```
