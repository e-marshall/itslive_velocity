```python
def calc_seasonal_means_for_dict_filtered(itslive_dict, term_type):
    df_ls = []
    for key in itslive_dict.keys():
        ds = itslive_dict[key]
        ds_gb = ds.groupby(ds.mid_date.dt.season).mean()
        ds_gb = ds_gb.reindex({'season':['DJF','MAM','JJA','SON']})
        rgi_id = key
        winter_mean = ds_gb.sel(season='DJF').v_filtered.mean().data
        spring_mean = ds_gb.sel(season='MAM').v_filtered.mean().data
        summer_mean = ds_gb.sel(season='JJA').v_filtered.mean().data
        fall_mean = ds_gb.sel(season='SON').v_filtered.mean().data
        term_type = term_type
        d = {'RGIId':rgi_id, 'winter': winter_mean, 'spring':spring_mean, 'summer':summer_mean,'fall':fall_mean, 'TermType':term_type}
        df = pd.DataFrame(d, index = [0])
        df_ls.append(df)
    df_full = pd.concat(df_ls)
    return df_full
```

```python
def calc_seasonal_means_for_dict_98(itslive_dict, term_type):
    '''take a dict of glacier velocity objects for just one term type (ie. all lake-term or all landterm glaciers) 
    groupby season and reindex season order for each glacier object, then calc seasonal mean speeds (averaged over area of ind glacier) and organize into a pd df
    '''
    df_ls = []
    for key in itslive_dict.keys():
        ds = itslive_dict[key]
        ds_gb = ds.groupby(ds.mid_date.dt.season).mean()
        ds_gb = ds_gb.reindex({'season':['DJF','MAM','JJA','SON']})
        rgi_id = key
        winter_mean = ds_gb.sel(season='DJF').vf_98.mean().data
        spring_mean = ds_gb.sel(season='MAM').vf_98.mean().data
        summer_mean = ds_gb.sel(season='JJA').vf_98.mean().data
        fall_mean = ds_gb.sel(season='SON').vf_98.mean().data
        term_type = term_type
        d = {'RGIId':rgi_id, 'winter': winter_mean, 'spring':spring_mean, 'summer':summer_mean,'fall':fall_mean, 'TermType':term_type}
        df = pd.DataFrame(d, index = [0])
        df_ls.append(df)
    df_full = pd.concat(df_ls)
    
    return df_full
```

```python
def seasonal_mean_df_wrapper(df1, df2):
    '''merge and re-organize seasonal mean glacier velocity dfs that are separated by term type. ie take a laketerm df and a landterm df, combine and organize for analysis (wide to long):
    '''
    df_full = pd.concat([df1, df2])
    df_long = df_full.melt(id_vars = ['RGIId','TermType'], value_vars = ['winter','spring','summer','fall']).rename({'variable':'season','value':'its_velocity'}, axis=1)
    
    return df_long
```

```python
def subset_itslive_by_sensor(input_dict, sensor):
    
    keys_ls, vals_ls = [],[]
    
    if sensor == 'S1':
        for key in input_dict.keys():
        
            keys_ls.append(key)
            s1_condition = input_dict[key].satellite_img1.isin(['1A','1B'])
            s1_subset = input_dict[key].sel(mid_date=s1_condition)
            vals_ls.append(s1_subset)
    elif sensor == 'S2':
        for key in input_dict.keys():
        
            keys_ls.append(key)
            s1_condition = input_dict[key].satellite_img1.isin(['2A','2B'])
            s1_subset = input_dict[key].sel(mid_date=s1_condition)
            vals_ls.append(s1_subset)
            
    elif sensor == 'L9':
        for key in input_dict.keys():
            keys_ls.append(key)
            l9_condition = input_dict[key].satellite_img1 == '9.'
            l9_subset = input_dict[key].sel(mid_date = l9_condition)
            vals_ls.append(l9_subset)
    
    elif sensor == 'L8':
        for key in input_dict.keys():
            keys_ls.append(key)
            l8_condition = input_dict[key].satellite_img1 == '8.'
            l8_subset = input_dict[key].sel(mid_date = l8_condition)
            vals_ls.append(l8_subset)
            
    elif sensor == 'L7':
        for key in input_dict.keys():
            keys_ls.append(key)
            l7_condition = input_dict[key].satellite_img1 == '7.'
            l7_subset = input_dict[key].sel(mid_date = l7_condition)
            vals_ls.append(l7_subset)
    
    subset_dict = dict(zip(keys_ls, vals_ls))
    
    return subset_dict

```

```python
def itslive_sensor_subset_seasonal_means_wrapper(input_dict, sensor, lake_ids_ls, land_ids_ls):
    '''currently hardcoded for MC filtered data, not 98th percentile 
    '''
    
    subset_dict = subset_itslive_by_sensor(input_dict, sensor)
    #now subset by termtype
    subset_lakes_dict = {key: subset_dict[key] for key in lake_ids_ls}
    subset_lands_dict = {key: subset_dict[key] for key in land_ids_ls}
    
    df_lands_filtered = calc_seasonal_means_for_dict_filtered(subset_lands_dict, 'land')
    df_lakes_filtered = calc_seasonal_means_for_dict_filtered(subset_lakes_dict, 'lake')
    
    df_full = seasonal_mean_df_wrapper(df_lands_filtered, df_lakes_filtered)
    
    return df_full
```

```python
def retreat_itslive_glacier_comparison(i_dict, r_dict, rgi_id, color):
    
    ax.set_title(f'comparison of ITSLIVE (blue), RETREAT (red) for {rgi_id}')
    ax.set_ylabel('magnitude of velocity (m/y)')
    ax.hlines(xmin = i_dict[rgi_id].acquisition_date_img1,
              xmax = i_dict[rgi_id].acquisition_date_img2,
              y = i_dict[rgi_id].vf_98.mean(dim=['x','y']), color=color)
    ax.hlines(xmin = r_dict[rgi_id].img1_date,
              xmax = r_dict[rgi_id].img2_date,
              y = r_dict[rgi_id].dis_mag_my.mean(dim=['x','y']), color='red')
```

```python
from matplotlib.patches import Patch

def compare_itslive_components_retreat_one_glacier(rgi_id):
    
    #ax.set_title(f'comparison of ITSLIVE velocities (by sensor) and RETREAT retreat velocities, {rgi_id}')

    colors = ['royalblue','mediumpurple','darkturquoise','darkcyan','red']
    labels = ['ITS-S1','ITS-S2','ITS-L8','ITS-L9','RETREAT']

    ax.hlines(xmin = s2_itslive_dict[rgi_id].acquisition_date_img1,
                  xmax = s2_itslive_dict[rgi_id].acquisition_date_img2,
                  y = s2_itslive_dict[rgi_id].vf_98.mean(dim=['x','y']), color='royalblue', label='S2')
    ax.hlines(xmin = s1_itslive_dict[rgi_id].acquisition_date_img1,
                  xmax = s1_itslive_dict[rgi_id].acquisition_date_img2,
                  y = s1_itslive_dict[rgi_id].vf_98.mean(dim=['x','y']), color='mediumpurple', label='S1')
    ax.hlines(xmin = l8_itslive_dict[rgi_id].acquisition_date_img1,
                  xmax = l8_itslive_dict[rgi_id].acquisition_date_img2,
                  y = l8_itslive_dict[rgi_id].vf_98.mean(dim=['x','y']), color='darkturquoise', label='L8')
    ax.hlines(xmin = l9_itslive_dict[rgi_id].acquisition_date_img1,
                  xmax = l9_itslive_dict[rgi_id].acquisition_date_img2,
                  y = l9_itslive_dict[rgi_id].vf_98.mean(dim=['x','y']), color='darkcyan', label='L9')
    ax.hlines(xmin = retreat_dict[rgi_id].img1_date,
                  xmax = retreat_dict[rgi_id].img2_date,
                  y = retreat_dict[rgi_id].dis_mag_my.mean(dim=['x','y']), color='red')
    
    legend_elements = [Patch(facecolor=colors[0], label = labels[0]),
                       Patch(facecolor=colors[1], label = labels[1]),
                       Patch(facecolor=colors[2], label = labels[2]),
                       Patch(facecolor=colors[3], label = labels[3]),
                       Patch(facecolor=colors[4], label=labels[4])]
    ax.legend(handles=legend_elements, loc='upper left')
```
