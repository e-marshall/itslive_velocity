```python
whole_valid = ds.dis_mag_my.count(dim=['x','y'])
whole_possible = ds.dis_mag_my.notnull().any('time').sum(['x','y'])
cov_whole = whole_valid/whole_possible
```

```python
area_indicator_upper = cov_upper/ cov_whole
area_indicator_lower = cov_lower/cov_whole
```

```python
r_key_ls, r_val_ls = [],[]

for f in range(len(retreat_files)):
    
    g = retreat_files[f][3:-3]
    r_key_ls.append(g)
    ds = xr.open_dataset(os.path.join(retreat_path, retreat_files[f]))
    r_val_ls.append(ds)
    
retreat_dict = dict(zip(r_key_ls, r_val_ls))       
```

```python
def retreat_seasonal_decomp(ds):
    
    decomp = seasonal_decompose(ds.dis_mag_my.mean(dim=['x','y']), model = 'multiplicative', period=6)
    trend = decomp.trend
    seasonal = decomp.seasonal
    residual = decomp.resid
    
    fig, axs = plt.subplots(ncols=3, figsize=(18,5))
    axs[0].set_title('trend')
    axs[1].set_title('seasonal')
    axs[2].set_title('residual')
    axs[0].plot(trend)
    axs[1].plot(seasonal)
    axs[2].plot(residual)
```

```python
r_decomposition_10255 = seasonal_decompose(r_test.dis_mag_my.mean(dim=['x','y']), model='multiplicative', period=6)
```

```python
trend = r_decomposition_10255.trend
seasonal = r_decomposition_10255.seasonal
residual = r_decomposition_10255.resid
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(15,5))
axs[0].plot(trend)
axs[1].plot(seasonal)
axs[2].plot(residual)
```
