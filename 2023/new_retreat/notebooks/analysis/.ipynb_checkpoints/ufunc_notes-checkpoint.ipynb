```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm
from statsmodels.stats.stattools import medcouple
import math
```

```python
#make toy data approximating data shape (1 glacier, RGI60-15.03976) x,y,time 
t = 2885
t_coord = np.arange(t)
x = 26
x_coord = np.arange(x)
y = 31
y_coord = np.arange(y)
numVals = (t*x*y)
maxVal = 150
skewness = 5

random = skewnorm.rvs(a=skewness, loc=maxVal, size=numVals)
random = random - min(random)
random = random / max(random)
random = random * maxVal
random_flat = random
random = np.reshape(random, (t,x,y))

#https://itecnote.com/tecnote/python-create-random-numbers-with-left-skewed-probability-distribution/
#https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
```

```python
rand_da = xr.DataArray(data = random,
                       coords = {'time':t_coord,
                                'x':x_coord,
                                'y':y_coord},
                       dims = ('time','x','y'))
```

```python
rand_da.dims
```

```python
rand_ds = xr.Dataset({'v':rand_da})
```

```python
mu = np.mean(random)
sigma = np.std(random)
count, bins, ignored = plt.hist(random_flat, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

looked like this: 

```python
#first reshaped data 
ds_vxy = rand_ds.v.stack(xy=('x','y'))
```

```python
#then wrote function to perform MedCouple reduction
def mc(arr, axis='xy'):
    #print(f'arr: {arr.shape}')
    
    arr = np.reshape(arr,(len(arr)))
    #print(arr.shape)
    s = pd.Series(arr).dropna()
    #print(s.shape)
    mc = medcouple(s)
    #print(mc)
    #print('----')
    
    return mc
```

```python
arr = np.reshape(arr, len(arr))
```

```python
def adj_boxplot(v): #passing in 2d np array
    '''function to calculate adjusted boxplot for skewed distributes (from Hubert + Vanderviernan 2008) 
    for outlier detection of velocity data
    '''
    #print(v.shape)
    v_stack = v.flatten() #flatten to 1d array
    #print(v_stack.shape)
    #print(type(v_stack))
    
    q1 = np.nanpercentile(v_stack, 25)
    q3 = np.nanpercentile(v_stack, 75) 
    iqr = (q3 - q1)
    
    arr = np.reshape(v_stack,(len(v_stack)))
    #print(arr.shape)
    #print(type(arr))
    s = pd.Series(v_stack).dropna()
    mc = medcouple(s)
    #print(np.array_equal(v_stack, arr))
    
    iqr = (q3-q1)
    if mc >= 0:
        lb = q1 - (1.5*math.exp(-4*mc)*iqr)
        ub = q3 + (1.5*math.exp(3*mc)*iqr)
    elif mc < 0:
        lb = q1 - (1.5*math.exp(-3*mc)*iqr)
        ub = q3 + (1.5*math.exp(4*mc)*iqr)
    
    filtered = np.where(np.logical_and((lb <= v),(v <= ub)), v, np.nan)
  
    return filtered
```

and vectorized using apply_ufunc():

```python
v_filtered = xr.apply_ufunc(adj_boxplot, rand_ds.v,
                        input_core_dims = [['x','y']], #passing in a 2d array (only 1 arg being passed)
                        output_core_dims =[['x','y']], #expecting a 2d array
                        exclude_dims = set(('x','y')), #we are just turning invalid pixels to nans, not changing the shape of the arr we pass, so no excluded dims
                        vectorize=True,
                                           )
```
